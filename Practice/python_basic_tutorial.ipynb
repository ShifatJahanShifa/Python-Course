## Python Lists

### Characteristics of python list
*   built-in data structure
*   stores ordered collection of items
*   dynamic data structure
*   can store data of multiple types
*   index-based starting from 0
*   stores references to objects not the actual values




```python
list1 = [1,2,3]
print(list1)

list2 = list(list1) #shallow copy, from another list
print(list2)

list3 = list('shifa') #from a string
print(list3)

list4 = list((1,2,3,'hi',6.7)) #from a tuple
print(list4)
```

```python
list4[-2]
```

```python
# add at the end of the list
# list4.append(5)
print(list4)

# add at a specific position
list4.insert(3,10)
print(list4)

# add multiple objects
list4.extend([11,12,13])
print(list4)
```

```python
list4.remove(5)
print(list4)

list4.pop()
print(list4)

list4.pop(4)
print(list4)
```

```python
for item in list4:
  print(item)
```

```python
list5 = [list1, list2, list3]
print(list5)
```

## Python Dictionary

### Characteristics of python dictionary


*   stores information in key-value pair
*   keys must be unique and immutable (string, number)
*   no type or mutability restriction on values

```python
dict1 = {
    "name": "shifa",
    "age": 25
}

print(dict1)
```

```python
dict2 = dict(name="shifa", age=25)
print(dict2)
```

```python
dict1['age']
```

```python
dict2.get('name') # safer version. in case of missing key, it returns None or a default value instead of error
```

```python
try:
  dict2['roll']
except:
  print("key not found")
```

```python
dict2.get('roll')
```

```python
dict2['roll'] = 1301 # adding
dict2['age'] = 24 # updating
print(dict2)
```

```python
del dict1['age']
dict1
```

```python
dict2.pop('roll')
dict2
```

```python
dict1.popitem()
```

```python
dict1
```

```python
for key in dict2:
  print(key)

for value in dict2.values():
  print(value)

for key, value in dict2.items():
  print(key, value)
```

```python
dict3 = {
    "student" : {
        "name": "shifa",
        "age": 25
    }
}

print(dict3)
print(dict3['student'])
print(dict3['student']['name'])
```

## Python Sets

### Characteristics of python sets
*   Collection of unique items
*   unordered; non index-based
*   fast search, insertion, deletion operations using hashing internally
*   does not sort the items as like in c++
*   can store heterogeneous elements
*   frozenset restricts the use of add() and remove() method with set
*   individual item is immutable



```python
set1 = {1,2,2,3}
set1
```

```python
type(set1)
```

```python
set2 = set([123456,314236547,57356899900000,23334])
set2
```

```python
set1.add(9)
set1
```

```python
set3 = set1.union(set2)
set3
```

```python
set4 = set2.intersection(set3)
set4
```

```python
set5 = set3.difference(set4)
set5
```

```python
set6 = {1,6,7}
set7 = set6.union(set5)
set7
```

```python
set8 = frozenset(set7)
set8
```

```python
set7.remove(6)
```

```python
# cannot perform add() or remove() operation to a frozenset.
# set8.add(8)
# set8.remove(9)
```

## Python Exception Handling

It increases program's reliability as it ensures that program does not terminate unexpectedly. There are four main keywords for handling exceptions: try, except, else and finally. Specific types of exceptions make the code safer and easier to debug.

```python
a = 10
b = 0
res = None

try:
  res = a/b
except ZeroDivisionError:
  print("can't divide by zero!")
else:
  print(res)
finally:
  print('done!')
```

```python
def check_age_validity(age):
  if age<0:
    raise ValueError("age can't be negative!")
  print("age is set.")

try:
  age = input("enter your age: ")
  age = int(age)
  check_age_validity(age)
except ValueError as e:
  print(e)
```
