```python
import numpy as np

students = [
    ('Ankit', 22, 'Up', 'Geu'),
    ('Ankita', np.nan, 'Delhi', np.nan),
    ('Rahul', 16, 'Tokyo', 'Abes'),
    ('Simran', 41, 'Delhi', 'Gehu'),
    ('Shaurya', np.nan, 'Delhi', 'Geu'),
    ('Shivangi', 35, 'Mumbai', np.nan),
    ('Swapnil', 35, np.nan, 'Geu'),
    (np.nan, 35, 'Uk', 'Geu'),
    ('Jeet', 35, 'Guj', 'Gehu'),
    (np.nan, np.nan, np.nan, np.nan)
]

details = pd.DataFrame(
    students,
    columns=['Name', 'Age', 'Place', 'College'],
    index=['a', 'b', 'c', 'd', 'e', 'f', 'g', 'i', 'j', 'k']
)

print("Boolean DataFrame:\n", details.isnull())
print("\nCount of NaN in each column:\n", details.isnull().sum())
print("\nCount total NaN in the DataFrame:\n", details.isnull().sum().sum())
```

```python
print(details.size)
print(details.shape)
print(details.ndim)
```

#### Creating a Dataframe to select rows with max and min values in Dataframe

```python
dictionary = {'Driver': ['Hamilton', 'Vettel', 'Raikkonen',
                    'Verstappen', 'Bottas', 'Ricciardo',
                    'Hulkenberg', 'Perez', 'Magnussen',
                    'Sainz', 'Alonso', 'Ocon', 'Leclerc',
                    'Grosjean', 'Gasly', 'Vandoorne',
                    'Ericsson', 'Stroll', 'Hartley', 'Sirotkin'],

         'Points': [408, 320, 251, 249, 247, 170, 69, 62, 56,
                    53, 50, 49, 39, 37, 29, 12, 9, 6, 4, 1],

         'Age': [33, 31, 39, 21, 29, 29, 31, 28, 26, 24, 37,
                 22, 21, 32, 22, 26, 28, 20, 29, 23]}

# creating dataframe using DataFrame constructor
df5 = pd.DataFrame(dictionary)
print(df5.head(10))
```

```python
# Example 1: Shows max on Driver, Points, and Age columns.
print(df5.max())
print()

# Example 2: Who scored max points
print(df5[df5.Points == df5.Points.max()])
print()

# Example 3: What is the maximum age
print(df5.Age.max())
print()

# Example 4: Who is the oldest driver
print(df5[df5.Age == df5.Age.max()])
```

```python
# Example 1: Shows min on Driver, Points, Age columns.
print(df5.min())
print()

# Example 2: Who scored min points
print(df5[df5.Points == df5.Points.min()])
print()

# Example 3: Which row has minimum age in the Dataframe who is the youngest driver
print(df5[df5.Age == df5.Age.min()])
```
