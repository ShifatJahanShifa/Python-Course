## Python NumPy

### Characteristics of python numpy


*   array processing package
*   provides multidimensional array object that stores elements of same data type
*   number of dimensions in an array = rank (row)
*   size of the array along each dimension = shape (column)




```python
import numpy as np

array1 = np.array([1,2,3,4,5]) # from list
print(array1)

array2 = np.array((1,2,3,4,5)) # from tuple
print(array2)

array3 = np.array([[1,2,3],[4,5,6]]) # 2D array
print(array3)
```

```python
array3.shape
```

```python
array3.size
```

```python
# numpy array behavior when array contains characters
array4 = np.array(['a','b'])
array4
```

```python
array5 = array1[2]
array5
```

```python
import numpy as np

array6 = np.array([
    [-1, 2, 0, 4],
    [4, -0.5, 6, 0],
    [2.6, 0, 7, 8],
    [3, -7, 4, 2.0]
])

print(array6)

# slicing
array7 = array6[::2, :3] # row wise and column wise
print(array7)

# access specific element
array8 = array6[1,2]
print(array8)

# access specific elements
array9 = array6[[1,1,2,3], [3,2,3,0]]
print(array9)
```

```python
a = np.array([[1, 2],
              [3, 4]])

b = np.array([[4, 3],
              [2, 1]])

print("Adding 1 to every element:\n", a + 1)
print("Subtracting 2 from each element:\n", b - 2)
print("Sum of all array elements:", a.sum())
print("Array sum:\n", a + b)
```

```python
array10 = np.array([1,3,4], dtype=np.int32)
array10
print(array10.dtype)
```

```python
print(np.add(a,b))
print(np.subtract(a,b))
print(np.multiply(a,b))
print(np.divide(a,b))
print(np.sqrt(a))
print(a.T)
```
